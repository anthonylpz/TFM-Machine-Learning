```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga de datos
data = pd.read_csv('BDD_ENE_17_ABR_24.csv', encoding='latin1')
```

```python
# Seleccionar solo las variables relevantes
columnas = [
    'ANIO', 'PROVINCIA', 'CANTON', 'FECHA', 'HORA', 'TIPO_DE_SINIESTRO',
    'TIPO_DE_VEHICULO_1', 'EDAD_1', 'SEXO_1', 'CONDICION_1', 'CASCO_1', 'CINTURON_1'
]
df = data[columnas]
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

# Agrupar por año y contar la cantidad de accidentes
accidentes_por_anio = df.groupby('ANIO').size()

# Crear el gráfico de líneas
plt.figure(figsize=(12, 6))
plt.plot(accidentes_por_anio.index, accidentes_por_anio.values, marker='o', linestyle='-', color='#2E86C1', linewidth=2)  # Cambio de color

# Personalizar el gráfico
plt.title('Evolución de la cantidad de accidentes de tránsito por año', fontsize=16, fontweight='bold', color='#424242')  # Cambio de color y estilo de fuente
plt.xlabel('Año', fontsize=14, fontweight='bold', color='#424242')  # Cambio de color y estilo de fuente
plt.ylabel('Cantidad de accidentes', fontsize=14, fontweight='bold', color='#424242')  # Cambio de color y estilo de fuente
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(accidentes_por_anio.index, rotation=45, fontsize=12, fontweight='bold', color='#424242')  # Cambio de color y estilo de fuente
plt.yticks(fontsize=12, fontweight='bold', color='#424242')  # Cambio de color y estilo de fuente
plt.tight_layout()

# Añadir etiquetas en los puntos de datos más arriba de la línea
for x, y in zip(accidentes_por_anio.index, accidentes_por_anio.values):
    plt.text(x, y + 20, f'{y}', ha='center', va='bottom', fontsize=10, fontweight='bold', color='#424242')  # Ajuste para colocar más arriba

# Estilizar el borde del gráfico
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ajustar el tamaño del gráfico
plt.figure(figsize=(12, 6))

# Crear el gráfico de barras utilizando seaborn
sns.countplot(x='TIPO_DE_VEHICULO_1', data=df, order=df['TIPO_DE_VEHICULO_1'].value_counts().index, palette='viridis')

# Personalizar el gráfico
plt.title('Número de accidentes según el tipo de vehículo', fontsize=16, fontweight='bold', color='#424242')  # Cambio de color y estilo de fuente
plt.xlabel('Tipo de vehículo', fontsize=14, fontweight='bold', color='#424242')  # Cambio de color y estilo de fuente
plt.ylabel('Cantidad de accidentes', fontsize=14, fontweight='bold', color='#424242')  # Cambio de color y estilo de fuente
plt.xticks(rotation=45, ha='right', fontsize=12, fontweight='bold', color='#424242')  # Alineación y estilo de fuentes para el eje x
plt.yticks(fontsize=12, fontweight='bold', color='#424242')  # Cambio de color y estilo de fuente para el eje y
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))  # Ajusta el tamaño del gráfico

# Graficar la distribución de accidentes por tipo de siniestro
sns.countplot(x='TIPO_DE_SINIESTRO', data=df, order=df['TIPO_DE_SINIESTRO'].value_counts().index)
plt.title('Distribución de accidentes por tipo de siniestro')
plt.ylabel('Cantidad de accidentes')

# Ajustar las etiquetas del eje x (vertical)
plt.xticks(rotation=45, ha='right')

plt.tight_layout()  # Ajusta automáticamente el espaciado del gráfico
plt.show()
```
